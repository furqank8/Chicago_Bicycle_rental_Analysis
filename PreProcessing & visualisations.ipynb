```r
membership_data <- rbind(
  bicycle_data %>%
  filter(member_casual == "casual") %>%
  group_by(year, month, day, start_station_name, member_casual) %>%
  summarise(total=n(), .groups = "drop") %>%
  mutate(new = c(NA, total[1:length(total) - 1])) %>%
  mutate(num = case_when(total - new <= 0 ~new,TRUE ~ total-new )), 
  bicycle_data %>%
  filter(member_casual == "member") %>%
  group_by(year, month, day, start_station_name, member_casual) %>%
  summarise(total=n(), .groups = "drop") %>%
  mutate(new = c(NA, total[1:length(total) - 1])) %>%
  mutate(num = case_when(total - new <= 0 ~new,TRUE ~ total-new )))


head(membership_data)
```

```r
membership_data = membership_data %>% select(-total, -new)
head(membership_data)
```

```r
membership_data_casual <- membership_data %>% filter(member_casual == "casual") %>%  select(-member_casual)
```

```r
station_names_vect <- membership_data_casual %>% pull(start_station_name)
membership_data_casual$start_station_name = factor(membership_data_casual$start_station_name, levels = station_names_vect, labels = c(1:length(station_names_vect)))


# Using sample_frac to create 70 - 30 split into test and train
train <- sample_frac(membership_data_casual, 0.7)
sample_id <- as.numeric(rownames(train))
test <- membership_data_casual[-sample_id,]

# Setting the baseline to "classic_bike"
#train$no_of_rides <- relevel(factor(train$no_of_rides), ref = "1")

# Training the multinomial model
lin_model <- glm(num ~ start_station_name + year + month + day , data = train)

# Checking the model
summary(lin_model)
```

```r
test_original = test$num
test = test %>% select(-num)

# Predicting the values for test dataset
test$ClassPredicted <- predict(lin_model, test)
test$original_data <- test_original

# Building confusion matrix
tab <- table(test_original, test$ClassPredicted)

# Calculating accuracy - sum of diagonal elements divided by total obs
round((sum(diag(tab))/sum(tab))*100,2)
```
